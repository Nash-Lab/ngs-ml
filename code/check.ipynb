```python
out = pd.read_csv('test.txt')
out
```

```python
out.assign(freq=lambda x: x.barcode.map(out['barcode'].value_counts())).query('found == 1').query('freq > 1').drop_duplicates()[['aa_mutation','freq']].groupby(['aa_mutation']).sum()
```

```python
out.query('found == 1').assign(freq=lambda x: x.barcode.map(out['barcode'].value_counts())).query('freq > 1').drop_duplicates()[['aa_mutation','freq']].fillna('wt').groupby(['aa_mutation'], as_index=False).sum()
```
